```python
import pandas as pd
import matplotlib.pyplot as plt
data1 = pd.read_csv('stock (2).csv')

data1['Dividend/share'] = data1['Dividend/share'].str.replace('$', '').str.replace('%', '').astype(float)
data1_sorted = data1.sort_values(by='Dividend/share', ascending=False)
top_5 = data1_sorted.head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5['Symbol'], top_5['Dividend/share'], color='skyblue')
plt.xlabel('Symbol')
plt.ylabel('Dividend/Share')
plt.title('5 Dividend/share tertinggi')
plt.xticks(rotation=45)

for i, (symbol, dividend) in enumerate(zip(top_5['Symbol'], top_5['Dividend/share'])):
    plt.text(i, dividend + 0.1, top_5.iloc[i]['Stock Name'], ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data1 = pd.read_csv('stock (2).csv')

data1['Market Value'] = data1['Market Value'].str.replace(',', '').str.replace('$', '')
def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        parts = value.split('.')
        if len(parts) > 2:
            return float('.'.join(parts[:2]) + parts[2])
        else:
            raise

data1['Market Value'] = data1['Market Value'].apply(convert_to_float)
data1_sorted = data1.sort_values(by='Market Value', ascending=False)
top_5 = data1_sorted.head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5['Symbol'], top_5['Market Value'], color='skyblue')
plt.xlabel('Symbol')
plt.ylabel('Market Value')
plt.title('5 Market Value tertinggi')
plt.xticks(rotation=45)

for i, (symbol, market_value) in enumerate(zip(top_5['Symbol'], top_5['Market Value'])):
    plt.text(i, market_value + 0.1, top_5.iloc[i]['Stock Name'], ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
